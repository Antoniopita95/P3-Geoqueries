### Filtrar ciudades

Despu√©s de habernos conoctado a la base de Mongo, vamos a realizar querys para filtrar unos ciertos requisitos que requiere la empresa. Para ello, he elegido 3 ciudades de la base de datos: Madrid, actual campeona de Europa con el equipo Mad Lions, San Francisco y Nueva York, siendo estas grandes representantes del sector en el continente americano

```python
proj = {"_id": 0, "name": 1, "offices.city":1, "number_of_employees": 1, "offices.latitude" : 1, "offices.longitude" : 1}
cond1 = {"category_code": "games_video"}
cond2 = {"$or": [{"offices.city": "Madrid"}, {"offices.city":"San Francisco"}, {"offices.city": "New York"}]}
cond3 = {"number_of_employees": {"$lt":120, "$gt":20}}
companies = list(collection.find({"$and": [cond1,cond2,cond3]}, proj))
df = pd.DataFrame(companies)
df
```
